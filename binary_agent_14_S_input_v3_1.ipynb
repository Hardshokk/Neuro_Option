```python
class NewAgent(Agent):
    def __init__(self):
        """отдельно"""
        self.path_dir = "width_variant_s_v3_1"
        self.symbol = "GBPCAD"
        self.tf = mt5.TIMEFRAME_M15
        self.qv_bars = 12
        # self.optimizer = Nadam(learning_rate=0.00001, beta_1=0.3, beta_2=0.9)
        self.optimizer = RMSprop(learning_rate=0.00001, rho=0.9)
        self.terminal = 0
        """конец отдельно"""
        Agent.__init__(self)
        
    def create_model(self):
        """input блок"""
        input_0 = Input((12, 4), batch_size=1, name='my_input_0')
        input_1 = Input((12, 1), batch_size=1, name='my_input_1')
        input_2 = Input((12, 1), batch_size=1, name='my_input_2')
        input_3 = Input((12, 1), batch_size=1, name='my_input_3')
        input_4 = Input((12, 1), batch_size=1, name='my_input_4')
        input_5 = Input((12, 1), batch_size=1, name='my_input_5')
        input_6 = Input(24, batch_size=1, name='my_input_6')
        input_7 = Input(4, batch_size=1, name='my_input_7')
        input_8 = Input((12, 1), batch_size=1, name='my_input_8')
        input_9 = Input((12, 1), batch_size=1, name='my_input_9')
        input_10 = Input((12, 1), batch_size=1, name='my_input_10')
        input_11 = Input((12, 1), batch_size=1, name='my_input_11')
        input_12 = Input((12, 1), batch_size=1, name='my_input_12')
        input_13 = Input((12, 1), batch_size=1, name='my_input_13')
        """блок 1"""
        lay0 = LSTM(36, activation='linear', recurrent_dropout=0.3)(input_0)
        lay0 = LeakyReLU()(lay0)
        lay1 = LSTM(36, activation='linear', recurrent_dropout=0.3)(input_1)
        lay1 = LeakyReLU()(lay1)
        lay2 = LSTM(36, activation='linear', recurrent_dropout=0.3)(input_2)
        lay2 = LeakyReLU()(lay2)
        lay3 = LSTM(36, activation='linear', recurrent_dropout=0.3)(input_3)
        lay3 = LeakyReLU()(lay3)
        lay4 = LSTM(36, activation='linear', recurrent_dropout=0.3)(input_4)
        lay4 = LeakyReLU()(lay4)
        lay5 = LSTM(36, activation='linear', recurrent_dropout=0.3)(input_5)
        lay5 = LeakyReLU()(lay5)
        lay6 = Dense(24, activation='linear')(input_6)
        lay6 = LeakyReLU()(lay6)
        lay7 = Dense(4, activation='linear')(input_7)
        lay7 = LeakyReLU()(lay7)
        lay8 = LSTM(36, activation='linear', recurrent_dropout=0.3)(input_8)
        lay8 = LeakyReLU()(lay8)
        lay9 = LSTM(36, activation='linear', recurrent_dropout=0.3)(input_9)
        lay9 = LeakyReLU()(lay9)
        lay10 = LSTM(36, activation='linear', recurrent_dropout=0.3)(input_10)
        lay10 = LeakyReLU()(lay10)
        lay11 = LSTM(36, activation='linear', recurrent_dropout=0.3)(input_11)
        lay11 = LeakyReLU()(lay11)
        lay12 = LSTM(36, activation='linear', recurrent_dropout=0.3)(input_12)
        lay12 = LeakyReLU()(lay12)
        lay13 = LSTM(36, activation='linear', recurrent_dropout=0.3)(input_13)
        lay13 = LeakyReLU()(lay13)
        """блок 2"""
        lay14 = Reshape((6, 8))(input_0)
        lay14 = LSTM(36, activation='linear', recurrent_dropout=0.3)(lay14)
        lay14 = LeakyReLU()(lay14)
        lay15 = Reshape((6, 2))(input_1)
        lay15 = LSTM(36, activation='linear', recurrent_dropout=0.3)(lay15)
        lay15 = LeakyReLU()(lay15)
        lay16 = Reshape((6, 2))(input_2)
        lay16 = LSTM(36, activation='linear', recurrent_dropout=0.3)(lay16)
        lay16 = LeakyReLU()(lay16)
        lay17 = Reshape((6, 2))(input_3)
        lay17 = LSTM(36, activation='linear', recurrent_dropout=0.3)(lay17)
        lay17 = LeakyReLU()(lay17)
        lay18 = Reshape((6, 2))(input_4)
        lay18 = LSTM(36, activation='linear', recurrent_dropout=0.3)(lay18)
        lay18 = LeakyReLU()(lay18)
        lay19 = Reshape((6, 2))(input_5)
        lay19 = LSTM(36, activation='linear', recurrent_dropout=0.3)(lay19)
        lay19 = LeakyReLU()(lay19)
        lay20 = Reshape((6, 2))(input_8)
        lay20 = LSTM(36, activation='linear', recurrent_dropout=0.3)(lay20)
        lay20 = LeakyReLU()(lay20)
        lay21 = Reshape((6, 2))(input_9)
        lay21 = LSTM(36, activation='linear', recurrent_dropout=0.3)(lay21)
        lay21 = LeakyReLU()(lay21)
        lay22 = Reshape((6, 2))(input_10)
        lay22 = LSTM(36, activation='linear', recurrent_dropout=0.3)(lay22)
        lay22 = LeakyReLU()(lay22)
        lay23 = Reshape((6, 2))(input_11)
        lay23 = LSTM(36, activation='linear', recurrent_dropout=0.3)(lay23)
        lay23 = LeakyReLU()(lay23)
        lay24 = Reshape((6, 2))(input_12)
        lay24 = LSTM(36, activation='linear', recurrent_dropout=0.3)(lay24)
        lay24 = LeakyReLU()(lay24)
        lay25 = Reshape((2, 6))(input_13)
        lay25 = LSTM(36, activation='linear', recurrent_dropout=0.3)(lay25)
        lay25 = LeakyReLU()(lay25)
        
        """блок 3"""
        lay26 = Flatten()(input_0)
        lay26 = Dense(128, activation='linear')(lay26)
        lay26 = LeakyReLU()(lay26)
        lay26 = Dropout(0.3)(lay26)
        lay26 = Dense(128, activation='linear')(lay26)
        lay26 = LeakyReLU()(lay26)
        
        lay27 = Flatten()(input_1)
        lay27 = Dense(128, activation='linear')(lay27)
        lay27 = LeakyReLU()(lay27)
        lay27 = Dropout(0.3)(lay27)
        lay27 = Dense(128, activation='linear')(lay27)
        lay27 = LeakyReLU()(lay27)
        
        lay28 = Flatten()(input_2)
        lay28 = Dense(128, activation='linear')(lay28)
        lay28 = LeakyReLU()(lay28)
        lay28 = Dropout(0.3)(lay28)
        lay28 = Dense(128, activation='linear')(lay28)
        lay28 = LeakyReLU()(lay28)
        
        lay29 = Flatten()(input_3)
        lay29 = Dense(128, activation='linear')(lay29)
        lay29 = LeakyReLU()(lay29)
        lay29 = Dropout(0.3)(lay29)
        lay29 = Dense(128, activation='linear')(lay29)
        lay29 = LeakyReLU()(lay29)
        
        lay30 = Flatten()(input_4)
        lay30 = Dense(128, activation='linear')(lay30)
        lay30 = LeakyReLU()(lay30)
        lay30 = Dropout(0.3)(lay30)
        lay30 = Dense(128, activation='linear')(lay30)
        lay30 = LeakyReLU()(lay30)
        
        lay31 = Flatten()(input_5)
        lay31 = Dense(128, activation='linear')(lay31)
        lay31 = LeakyReLU()(lay31)
        lay31 = Dropout(0.3)(lay31)
        lay31 = Dense(128, activation='linear')(lay31)
        lay31 = LeakyReLU()(lay31)
        
        lay32 = Flatten()(input_8)
        lay32 = Dense(128, activation='linear')(lay32)
        lay32 = LeakyReLU()(lay32)
        lay32 = Dropout(0.3)(lay32)
        lay32 = Dense(128, activation='linear')(lay32)
        lay32 = LeakyReLU()(lay32)
        
        lay33 = Flatten()(input_9)
        lay33 = Dense(128, activation='linear')(lay33)
        lay33 = LeakyReLU()(lay33)
        lay33 = Dropout(0.3)(lay33)
        lay33 = Dense(128, activation='linear')(lay33)
        lay33 = LeakyReLU()(lay33)
        
        lay34 = Flatten()(input_10)
        lay34 = Dense(128, activation='linear')(lay34)
        lay34 = LeakyReLU()(lay34)
        lay34 = Dropout(0.3)(lay34)
        lay34 = Dense(128, activation='linear')(lay34)
        lay34 = LeakyReLU()(lay34)
        
        lay35 = Flatten()(input_11)
        lay35 = Dense(128, activation='linear')(lay35)
        lay35 = LeakyReLU()(lay35)
        lay35 = Dropout(0.3)(lay35)
        lay35 = Dense(128, activation='linear')(lay35)
        lay35 = LeakyReLU()(lay35)
        
        lay36 = Flatten()(input_12)
        lay36 = Dense(128, activation='linear')(lay36)
        lay36 = LeakyReLU()(lay36)
        lay36 = Dropout(0.3)(lay36)
        lay36 = Dense(128, activation='linear')(lay36)
        lay36 = LeakyReLU()(lay36)
        
        lay37 = Flatten()(input_13)
        lay37 = Dense(128, activation='linear')(lay37)
        lay37 = LeakyReLU()(lay37)
        lay37 = Dropout(0.3)(lay37)
        lay37 = Dense(128, activation='linear')(lay37)
        lay37 = LeakyReLU()(lay37)
        
        """блок объединения"""
        dense = Concatenate()([lay0, lay1, lay2, lay3, lay4, lay5, lay6, lay7, lay8, lay9, lay10, lay11, lay12, lay13,
                               lay14, lay15, lay16, lay17, lay18, lay19, lay20, lay21, lay22, lay23, lay24, lay25,
                               lay26, lay27, lay28, lay29, lay30, lay31, lay32, lay33, lay34, lay35, lay36, lay37])
        """output блок"""
        dense = Dense(1000)(dense)
        dense = Dropout(0.3)(dense)
        dense = Dense(1000)(dense)
        dense = Dropout(0.3)(dense)
        dense = Dense(500)(dense)
        dense = Dense(1, activation='sigmoid', name='my_output_1')(dense)

        return [input_0, input_1, input_2, input_3, input_4, input_5, input_6, input_7, input_8, input_9, input_10,
                input_11, input_12, input_13], [dense]
    
agent = NewAgent()
```
